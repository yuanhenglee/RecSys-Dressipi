```python
def get_X_y( file ):
    with open(file) as f:
        data = f.readlines()
    X = []
    y = []
    for line in data:
        row = [float(v) for v in line.split(',')]
        X.append(row[:-1])
        y.append(row[-1])
    return X, y

def get_X( file ):
    with open(file) as f:
        data = f.readlines()
    X = []
    for line in data:
        row = [float(v) for v in line.split(',')]
        X.append(row)
    return X
```

```python


# clf = RandomForestClassifier(max_depth=2, random_state=0)
clf = LogisticRegression(class_weight = 'balanced')
X_test, y_test = get_X_y( project_root_path + '/dataset/train_features/train_' + str(8) )

# fit training data
for i in range(1):
    start_time = time.time()
    file = project_root_path + '/dataset/train_features/train_' + str(i)
    X, y = get_X_y( file )
    clf.fit(X, y)
    print(file, "Done. Execution Time:", time.time() - start_time)
```

```python
clf.score(X_test, y_test)
```

```python
filename = project_root_path + '/model/logistic.model'
pickle.dump(clf, open(filename, 'wb'))
```
