```python
def get_similarity( file, start = 0 ):
    with open(file) as f:
        data = f.readlines()
    similarities = []
    purchased_id = 0
    # for line in data:
    for i in range(4990):
      line = data[start+i]
      row = [float(v) for v in line.split(',')]
      # just innerproduct
      similarities.append(row[0])
      if row[-1] == 1:
        purchased_id = i

    return similarities, purchased_id
```

```python
session_ids = []
with open( project_root_path + 'dataset/test_leaderboard_sessions.csv') as f:
  for line in f.readlines()[1:]:
    # if len(line):
    session_ids.append(int(line.split(',')[0]))
# session_ids
```

```python
candidate_ids = []
with open( project_root_path + 'dataset/candidate_items.csv') as f:
  for line in f.readlines()[1:]:
    # if len(line):
    candidate_ids.append(int(line))
```

```python
# train validate
for i in range(1):
  similarities, purchased_id = get_similarity( project_root_path + 'dataset/train_features/train_' + str(1), i*4990 )
  # print(similarities[purchased_id:purchased_id+10])
  if purchased_id != 0:
    # predict_rank = 4990-list(np.argsort( similarities )).index(purchased_id)
    print(list(np.argsort( similarities ))[-1:-101:-1])
    # print(predict_rank)
```

```python
result = []
for session_id in session_ids:
  similarities, purchased_id = get_similarity( project_root_path + 'dataset/test_features/test_s_' + str(session_id) )
  top_100_index = list(np.argsort( similarities ))[-1:-101:-1]
  for i in range(100):
    result.append(str(session_id) + ',' + str(candidate_ids[top_100_index[i]]) + ',' + str(i+1) + '\n')
```

```python
with open( project_root_path + 'result/VSM_ranking_LB.csv') as f:
  f.write("session_id,item_id,rank\n")
  f.write(''.join(result))
```
